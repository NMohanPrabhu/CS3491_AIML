```python
import pandas as pd
from pgmpy.models import BayesianNetwork
```

```python
#Load the dataset from your location
df = pd.read_csv('E:/Pendrive H/AIML_LAB_IT/4_BAYES_NETWORK/Patient_details.csv')
```

```python
df
```

```python
#Selecting necessary columns in the dataset
df1=df[['SR_2 Folder thickness(mm)', 'SR_3 Folder thickness(mm)','SR_4 Folder thickness(mm)','COVID-19 Infection']]
```

```python
#Renaming the columns and removing any null data in the dataset
dict = {'SR_2 Folder thickness(mm)': 'SR_2','SR_3 Folder thickness(mm)': 'SR_3','SR_4 Folder thickness(mm)': 'SR_4','COVID-19 Infection': 'Infection'}
df1.rename(columns=dict,inplace=True)
result= {"Positive": 1,"Negative":0}
df1['Infection']=df1['Infection'].map(result)
df1.isnull()
df1.dropna
```

```python
df1
```

```python
model=BayesianNetwork([("SR_2", "SR_3"),("SR_3", "SR_4"),("SR_2", "Infection"),("SR_3", "Infection"),("SR_4", "Infection")])
model.fit(df1)
```

```python
from pgmpy.inference import VariableElimination
infer=VariableElimination(model)
q1=infer.query(variables=['Infection'],evidence={'SR_2':4})
print(q1)
```
